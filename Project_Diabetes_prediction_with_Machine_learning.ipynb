# Diabetes Prediction With Machine Learning


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

Dbd = pd.read_csv('pima-indians-diabetes.csv')
Dbd
```

```python
Dbd.columns
```

```python
Dbd_copy
```

## Replace Every Lost Value In The Dataset With "Nan" So They can be Viewed as Null Values

```python
Dbd_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']] = Dbd_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']].replace(0,np.nan)
```

```python
Dbd_copy.isnull().sum()
```

## Then Replace The "Null Values" with The "mean" of each column

```python
#This Makes our data very clean to proceed

Dbd['Glucose'] = Dbd['Glucose'].replace(0,Dbd['Glucose'].mean())
Dbd['BloodPressure'] = Dbd['BloodPressure'].replace(0,Dbd['BloodPressure'].mean())
Dbd['SkinThickness'] = Dbd['SkinThickness'].replace(0,Dbd['SkinThickness'].mean())
Dbd['Insulin'] = Dbd['Insulin'].replace(0,Dbd['Insulin'].mean())
Dbd['BMI'] = Dbd['BMI'].replace(0,Dbd['BMI'].mean())
# Dbd.isnull().sum()
Dbd
```

## Overview of what our model is looking for to make prediction

```python
Dbd.groupby('Class').mean()
```

## Store the Feature variables in X and the Target(response) in y

```python
X = Dbd.drop('Class',axis=1)
y = Dbd['Class']
X
```

## Split the dataset

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

## Scikit Learn Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.pipeline import Pipeline
```

```python
#Creating pipelines for the whole algorithms that needs it

pipeline_lr = Pipeline([('scalar1',StandardScaler()),('lr_classifier',LogisticRegression())])
pipeline_knn = Pipeline([('scalar2',StandardScaler()),('knn_classifier',KNeighborsClassifier())])
pipeline_svm = Pipeline([('scalar3',StandardScaler()),('svm_classifier',SVC())])

pipeline_dt = Pipeline([('dt_classifier',DecisionTreeClassifier())])
pipeline_rf = Pipeline([('rf_classifier',RandomForestClassifier(max_depth=3))])
pipeline_gbc = Pipeline([('gbc_classifier',GradientBoostingClassifier())])
```

pipelines = [pipeline_lr,pipeline_knn,pipeline_svm,pipeline_dt,pipeline_rf,pipeline_gbc]
pipelines

```python
for pipe in pipelines:
    pipe.fit(X_train,y_train)
```

```python
pipe_dict = {0:'LR',1:'KNN',2:'SVC',3:'DT',4:'RF',5:'GBC'}
pipe_dict
```

```python
for i,model in enumerate(pipelines):
    print('{} Test Accuracy :{} '.format(pipe_dict[i],model.score(X_test,y_test)*100))
```

## Using our Best model to predict the outcome

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X,y)
```

## Create a new data for Prediction

```python
new_data = pd.DataFrame({
    'Pregnancies':6,
    'Glucose':148.0,
    'BloodPressure':72.0,
    'SkinThickness':35.0,
    'Insulin':79.799479,
    'BMI':33.6,
    'DiabetesPedigreeFunction':0.627,
    'Age':50,
},index=[0])
```

```python
p = model_lr.predict(new_data)
if p[0]==0:
    print('Non-Diabetic')
else:
    print('Diabetic')
```

## Save Model Using Joblib

```python
import joblib

joblib.dump(model_lr,'joblib_Lr_modelDiabetes')
```

## Import the model and try new prediction

```python
model = joblib.load('joblib_Lr_modelDiabetes')
p = model.predict(new_data)
if p[0]==0:
    print('Non-Diabetic')
else:
    print('Diabetic')
```
